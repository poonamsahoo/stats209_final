create confidence interval from the other data, see how many times the new data lies within the confidence interval???

```python
ADE_rep['std'] = list(df[['0.25 ADE', '0.5 ADE', '0.75 ADE']].std())
ADE_rep['lower'] = ADE_rep[0] - 1.96 * ADE_rep['std']
ADE_rep['higher'] = ADE_rep[0] + 1.96 * ADE_rep['std']
ADE_rep
```

```python
AIE_rep['std'] = list(df[['0.25 AIE', '0.5 AIE', '0.75 AIE']].std())
AIE_rep['lower'] = AIE_rep[0] - 1.96 * AIE_rep['std']
AIE_rep['higher'] = AIE_rep[0] + 1.96 * AIE_rep['std']
AIE_rep
```
