```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
#set parameters
vertices = 1000
adjacencies = 100
pi_vals = [0.25, 0.5, 0.75]
num_trials = 1000
```

```python
# AMY's SCRATCH

ADE = []
AIE = []
for i in range(50):
  G = nx.random_regular_graph(adjacencies, vertices)
  #A = nx.adjacency_matrix(G).todense()
  A = np.ones((vertices, vertices)) - np.eye(vertices)
  W = np.random.binomial(1, 0.25, vertices)
  noise = np.random.normal(0, 1, vertices)
  # Y = (W @ A)/200 + W * (1/2) + noise
  Y = (W @ A)/(3 *  np.sum(A, axis=1)) + W * (2/3) + noise
  average_direct_effect = (np.dot(Y,W))/sum(W) - (np.dot(Y, (1-W)))/sum((1-W))
  # treated = np.trace((A) @ np.outer(W, Y)) / sum(W)
  # untreated = np.trace((A) @ np.outer((1 - W), Y)) / sum(1 - W) # A - np.eye(vertices) # 1 - A
  # average_indirect_effect = treated - untreated
  average_indirect_effect = np.dot(Y, A @ W / sum(W) - A @ (1 - W) / sum(1 - W))
  ADE.append(average_direct_effect)
  AIE.append(average_indirect_effect)

print(sum(ADE)/len(ADE))
print(sum(AIE)/len(AIE))
```
