```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
G = nx.random_regular_graph(100, 1000)
pos = nx.spring_layout(G)
nx.draw(G, pos, node_size=10, with_labels=False, node_color='blue', edge_color='black', alpha=0.5, width=0.1)
plt.show()
```
